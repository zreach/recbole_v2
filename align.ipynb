```python
import os

def get_mp3_filenames(root_dir):
    mp3_filenames = []
    # 遍历根目录及所有子目录
    for dirpath, dirnames, filenames in os.walk(root_dir):
        for filename in filenames:
            # 检查文件是否以 .mp3 结尾（忽略大小写）
            if filename.lower().endswith('.mp3'):
                # 去除文件扩展名，获取前缀
                name_without_ext = os.path.splitext(filename)[0]
                mp3_filenames.append(name_without_ext)
    return mp3_filenames

# 示例使用

target_directory = "/data2/zhouyz"
mp3_list = get_mp3_filenames(target_directory)
# print(mp3_list)
```
